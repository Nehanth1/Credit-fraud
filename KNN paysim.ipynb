```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
df = pd. read_csv("C:\\Users\\Nehanth\\Desktop\\DSP\\project\\archive\\paysim1.csv")
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.head()
```

```python
paysim_data_copy = df.copy(deep = True)
paysim_data_copy[['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']] = paysim_data_copy[['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']].replace(0,np.NaN)
```

```python
paysim_data_copy['amount'].fillna(paysim_data_copy['amount'].mean(), inplace = True)
paysim_data_copy['oldbalanceOrg'].fillna(paysim_data_copy['amount'].mean(), inplace = True)
paysim_data_copy['newbalanceOrig'].fillna(paysim_data_copy['amount'].mean(), inplace = True)
paysim_data_copy['oldbalanceDest'].fillna(paysim_data_copy['amount'].mean(), inplace = True)
paysim_data_copy['newbalanceDest'].fillna(paysim_data_copy['amount'].mean(), inplace = True)
```

```python
paysim_data_copy.head()
```

```python
paysim_data_copy.corr()['isFraud']
```

```python
X = paysim_data_copy.iloc[:,[2,3,4,5,6]]
y = paysim_data_copy.iloc[:,[7]]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 1/3, random_state=42)
```

```python
%%time
knn = KNeighborsClassifier(n_neighbors=19, algorithm = "ball_tree")
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames = ['True'], colnames = ['Predicted'], margins = True)
```
